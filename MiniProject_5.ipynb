```python
# import libraries

# for data analysis
import pandas as pd
import numpy as np

# for data visualization
import plotly.express as px
import plotly.graph_objects as go
```

Import Data

```python
# import csv for covid-19 data

df_cov = pd.read_csv("https://raw.githubusercontent.com/justinjiholee/covid-trend/main/Covid%20Data.csv") 
```

```python
# preview data

df_cov.tail()
```

```python
# data for select countries

df_cov_china = df_cov[df_cov["location"].str.contains("China")]
df_cov_india = df_cov[df_cov["location"].str.contains("India")]
df_cov_korea = df_cov[df_cov["location"].str.contains("Korea")]
df_cov_canada = df_cov[df_cov["location"].str.contains("Canada")]
df_cov_turkey = df_cov[df_cov["location"].str.contains("Turkey")]
```

```python
df_cov_china.head()
```

```python
# clean data
df_covid_china = df_cov_china.iloc[:,[2,3,4,5,34,37,39,44]]
df_covid_india = df_cov_india.iloc[:,[2,3,4,5,34,37,39,44]]
df_covid_korea = df_cov_korea.iloc[:,[2,3,4,5,34,37,39,44]]
df_covid_canada = df_cov_canada.iloc[:,[2,3,4,5,34,37,39,44]]
df_covid_turkey = df_cov_turkey.iloc[:,[2,3,4,5,34,37,39,44]]
```

Data Merge

```python
df_covid_merge =  pd.concat([df_covid_china, df_covid_india, df_covid_korea, df_covid_canada, df_covid_turkey], axis = 0, sort = False)
```

```python
# preview merged data
df_covid_merge.head()
```

```python
df_covid_apr = df_covid_merge[df_covid_merge["date"]== "4/28/21"]
```

```python
df_covid_apr.head()
```

```python
# pivot table

df_pivot_covid = pd.pivot_table(df_covid_merge,
                                 index = ["location"],
                                 values = ["new_cases", "new_vaccinations", "people_vaccinated"],
                                 aggfunc = np.sum).reset_index()
```

Data Visualization

```python
import plotly.figure_factory as ff
```

```python
location = ["China", "India", "South Korea", "Canada", "Turkey"]

bar_covid = go.Figure(data=[
            go.Bar(name = "Vacinnations per Population", x = location, y = df_covid_apr["total_vaccinations_per_hundred"])
            ])
           

bar_covid.update_layout(title = "Vaccinations per Population",
            xaxis_title = "Country",
            yaxis_title = "Percentage(%)",
            height = 800,
            width = 500)
```

```python
fig = go.Figure([
    
    go.Scatter(
        name='China',
        x = df_covid_china['date'],
        y=df_covid_china['new_vaccinations'],
        mode='markers+lines',
        marker =dict(color='red', size=2),
        showlegend=True
    ),

    go.Scatter(
        name='India',
        x=df_covid_india['date'],
        y=df_covid_india['new_vaccinations'],
        mode='markers+lines',
        marker = dict(color='blue', size=2),
        showlegend=True
    ),

go.Scatter(
        name='South Korea',
        x=df_covid_korea['date'],
        y=df_covid_korea['new_vaccinations'],
        mode='markers+lines',
        marker=dict(color='yellow', size=2),
        showlegend=True
    ),

    go.Scatter(
        name='Canada',
        x=df_covid_canada['date'],
        y=df_covid_canada['new_vaccinations'],
        mode='markers+lines',
        marker=dict(color='green', size=2),
        showlegend=True
    ),

    go.Scatter(
        name='Turkey',
        x=df_covid_turkey['date'],
        y=df_covid_turkey['new_vaccinations'],
        mode='markers+lines',
        marker=dict(color='purple', size=2),
        showlegend=True
    ),
])
fig.update_layout(
    yaxis_title='New Vaccinations',
    title= 'Vaccinations Trend',
    hovermode= 'x')

```
